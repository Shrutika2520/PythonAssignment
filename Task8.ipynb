# Q-1. Write an SQL query to fetch intersecting records of two tables.

select worker.Id, worker.FIRST_NAME, workerclone.Id, workerclone.FIRST_NAME from worker inner join workerclone on workerclone.FIRST_NAME = worker.FIRST_NAME;

# Q-2. Write an SQL query to show records from one table that another table does not have.

SELECT w.Id, w.FIRST_NAME , wc.Id, wc.FIRST_NAME FROM workerclone w LEFT JOIN worker wc ON wc.FIRST_NAME = w.FIRST_NAME WHERE wc.FIRST_NAME is NULL;
