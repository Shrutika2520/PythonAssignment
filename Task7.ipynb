# Q-2. Write an SQL query to clone a new table from another table.

CREATE TABLE workerclone LIKE worker;
INSERT INTO workerclone (Id,FIRST_NAME,LAST_NAME,EMAIL,DEPARTMENT,SALARY,JOINING_DATE) SELECT Id,FIRST_NAME,LAST_NAME,EMAIL,DEPARTMENT,SALARY,JOINING_DATE FROM worker;
select * from workerclone;
