# Task_10

# Q-1. Write an SQL query to determine the 8th highest salary without using TOP or limit method.

SELECT SALARY FROM worker W1 WHERE 7 = ( SELECT COUNT( DISTINCT ( W2.SALARY ) ) FROM worker W2 WHERE W2.SALARY >= W1.SALARY );

# Q-2. Write an SQL query to fetch the list of employees with the same salary.

elect distinct W.FIRST_NAME, W.SALARY from worker W, worker W1 where W.SALARY = W1.SALARY and W.FIRST_NAME != W1.FIRST_NAME;
